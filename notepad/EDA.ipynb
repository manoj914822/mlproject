```python
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'str']  # numeric
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'str']  # categorical

print('we have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nwe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
df.head()
```

```python
df["total_score"] = df["math score"] + df["reading score"] + df["writing score"]
df["average_score"] = df["total_score"] / 3
df.head()
```

```python
reading_avg_score = df[df['reading score'] < 56]['average_score'].mean()

math_avg_score = df[df['math score'] < 25]['average_score'].mean()

print("Average score of students who scored less than 56 in reading:", reading_avg_score)

print("Average score of students who scored less than 25 in math:", math_avg_score)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
sns.histplot(df['math score'], bins=20, kde=True, ax=axs[0])
axs[0].set_title('Distribution of Math Scores')
sns.histplot(df['reading score'], bins=20, kde=True, ax=axs[1])
axs[1].set_title('Distribution of Reading Scores')  
sns.histplot(df['writing score'], bins=20, kde=True, ax=axs[2])
axs[2].set_title('Distribution of Writing Scores')
plt.tight_layout()
plt.show()
```
